```python
ds = xr.open_dataset('atmos_monthly.nc', decode_times=False)
ds
```

```python
ds = ds.assign_coords(lon=(((ds.lon + 180) % 360) - 180)).sortby('lon')
```

```python
ds.temp  # (time, pfull, lat, lon)
temp_mean = ds.temp.mean(dim="pfull")  # Now temp_mean has shape: (time, lat, lon)
```

```python
# Find the index where pfull is closest to 1 bar (1000 hPa)
pfull_1bar_index = (abs(ds.pfull - 1000)).argmin()  # index closest to 1 bar

# Select temperature data at 1 bar (1000 hPa) for all time slices, latitudes, and longitudes
temp_1bar = ds.temp[:, pfull_1bar_index, :, :]  # shape: (time, lat, lon)

# Create subplots for each time slice
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

# Loop over the 6 time slices
for i in range(6):
    temp_slice = temp_1bar.isel(time=i)  # select data for this time slice at 1 bar

    # Transpose temp_slice if needed (e.g., (lat, lon) format for plotting)
    temp_plot = temp_slice.transpose('lat', 'lon')

    # Plot temperature at 1 bar for the current time slice
    im = axes[i].pcolormesh(ds.lon, ds.lat, temp_plot, shading='auto', cmap='jet')
    axes[i].set_title(f"Temperature at 1 bar – Time {ds.time.values[i]}")
    axes[i].set_xlabel("Longitude")
    axes[i].set_ylabel("Latitude")

# Add colorbar and title
fig.suptitle("Temperature at 1 Bar (0-3000 hPa)", fontsize=16)
plt.tight_layout()
plt.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.04)
plt.show()
```

```python
# Convert hPa to bar and find the index closest to 30 mbar (3000 hPa)
target_pressure = 30  # mbar
target_hPa = target_pressure * 10  # convert to hPa

# Find index closest to target pressure
closest_idx = abs(ds.pfull - target_hPa).argmin().item()
print(f"Closest pressure: {ds.pfull.values[closest_idx]} hPa")

# Slice temperature at this level
temp_slice = ds.temp[:, closest_idx, :, :]  # shape: (time, lat, lon)

fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

for i in range(6):
    t = temp_slice.isel(time=i)
    
    # Transpose to match lat vs lon
    temp_plot = t.transpose('lat', 'lon')
    
    im = axes[i].pcolormesh(ds.lon, ds.lat, temp_plot, shading='auto', cmap='plasma', vmin=800, vmax=1600)
    axes[i].set_title(f"Temp at {ds.time.values[i]} days – {ds.pfull.values[closest_idx]/1000:.2f} bar")
    axes[i].set_xlabel("Longitude")
    axes[i].set_ylabel("Latitude")

# Colorbar
fig.suptitle(f"Temperature Distribution at {target_pressure} mbar", fontsize=16)
plt.tight_layout()
cbar = plt.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.04)
cbar.set_label("Temperature (K)")
plt.show()
```

```python
u_zonal_mean = ds.ucomp.mean(dim="lon")  # (time, pfull, lat)

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

for i in range(6):
    u_slice = u_zonal_mean.isel(time=i)  # (pfull, lat)

    # Transpose so axes are (lat, pressure)
    u_plot = u_slice.transpose('pfull', 'lat')  # shape = (30, 320)

    im = axes[i].contourf(ds.lat, ds.pfull, u_plot, levels=21, cmap='jet', extend='both') # Blue (westward winds) at the equator and Red (eastward winds) at mid-latitudes
    axes[i].set_title(f"Zonal Wind – Time {ds.time.values[i]}")
    axes[i].set_xlabel("Latitude")
    axes[i].set_ylabel("Pressure (hPa)")
    axes[i].invert_yaxis()  # Higher pressure at bottom
    axes[i].set_ylim(ds.pfull.max(), ds.pfull.min())

# Add colorbar
fig.suptitle("Zonal Mean Zonal Wind Profiles", fontsize=16)
plt.tight_layout()
plt.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.04)
plt.show()
```

```python
# Choose pressure level index (e.g., closest to 200 hPa)
plev_index = 5
plev_value = ds.pfull.values[plev_index]

# Select zonal wind at that pressure level
u_upper = ds.ucomp.isel(pfull=plev_index)  # shape: (time, lat, lon)

# Plot for each time slice
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

for i in range(6):
    u_plot = u_upper.isel(time=i)

    im = axes[i].contourf(ds.lon, ds.lat, u_plot, levels=21, cmap='jet', extend='both')
    axes[i].set_title(f"Jet Stream – {plev_value:.0f} hPa – Time {ds.time.values[i]}")
    axes[i].set_xlabel("Longitude")
    axes[i].set_ylabel("Latitude")

# Colorbar and layout
fig.suptitle(f"Jet Stream (Zonal Wind) at {plev_value:.0f} hPa", fontsize=16)
plt.tight_layout()
plt.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.04)
plt.show()
```

```python

# Convert pressure to mbar (same as hPa)
pfull = ds.pfull  # shape (30,)
pfull_hPa = pfull  # already in hPa, assuming your dataset is in hPa

# Pressure levels (in hPa) to extract
target_pressures = [1, 30, 1000]  # 1 mbar, 30 mbar, 1 bar
labels = ["1 mbar", "30 mbar", "1 bar"]

# Find nearest pressure levels
pressure_indices = [int(np.abs(pfull_hPa - p).argmin()) for p in target_pressures]

# Choose one time slice to show (e.g., last one)
time_index = -1  # or 0, if you prefer early time

# Prepare figure
fig, axes = plt.subplots(1, 3, figsize=(18, 5), constrained_layout=True)
for ax, p_idx, label in zip(axes, pressure_indices, labels):
    # Extract data
    temp_slice = ds.temp.isel(time=time_index, pfull=p_idx)
    u_slice = ds.ucomp.isel(time=time_index, pfull=p_idx)
    v_slice = ds.vcomp.isel(time=time_index, pfull=p_idx)
    
    lon, lat = ds.lon, ds.lat
    Lon, Lat = np.meshgrid(lon, lat)

    # Plot temperature
    im = ax.pcolormesh(lon, lat, temp_slice, shading='auto', cmap='jet')
    
    # Wind arrows (downsampled for clarity)
    stride = 15  # adjust for spacing
    ax.quiver(
        lon[::stride], lat[::stride],
        u_slice[::stride, ::stride], v_slice[::stride, ::stride],
        color='white', scale=500
    )

    ax.set_title(f"Temp + Wind @ {label}")
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    ax.set_xlim(0, 360)
    ax.set_ylim(-90, 90)

# Add colorbar
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), orientation='vertical', fraction=0.025, pad=0.04)
cbar.set_label("Temperature (K)")

plt.suptitle("Temperature (K) and Wind Vectors at Selected Pressure Levels", fontsize=16)
plt.show()
```

```python
# Convert pfull to bars and find the index closest to 1 bar
pfull_bars = ds.pfull / 1000
target_pressure = 1.0  # bar
closest_idx = abs(pfull_bars - target_pressure).argmin().item()

# Select the last time slice
last_time = ds.time[-1]

# Extract temperature at that time and pressure
temp_slice = ds.temp.sel(time=last_time).isel(pfull=closest_idx)  # shape: (lat, lon)

# Create masks for dayside (-90° to +90°) and nightside (other longitudes)
dayside = temp_slice.sel(lon=ds.lon[(ds.lon >= -90) & (ds.lon <= 90)])
nightside = temp_slice.sel(lon=ds.lon[(ds.lon < -90) | (ds.lon > 90)])

# Compute zonal (longitude) mean temperature for each latitude
mean_dayside = dayside.mean(dim='lon')
mean_nightside = nightside.mean(dim='lon')

# Plot the contrast
plt.figure(figsize=(8, 6))
plt.plot(mean_dayside, ds.lat, label='Dayside', color='orange')
plt.plot(mean_nightside, ds.lat, label='Nightside', color='blue', linestyle=':')
plt.xlabel("Temperature (K)")
plt.ylabel("Latitude")
plt.title(f"Day–Night Zonal Mean Temperature at ~{pfull_bars[closest_idx].values:.2f} bar\nTime: {last_time.values}")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get last time index
last_time = ds.time[-1]

# Select the temperature slice at the closest pressure level (~1 bar) and last time
temp_slice = ds.temp.sel(time=last_time, pfull=pfull[closest_idx])

# Calculate temperature anomaly (remove global mean)
temp_anomaly = temp_slice - temp_slice.mean(dim=['lat', 'lon'])

# Plot the anomaly
plt.figure(figsize=(10, 5))
im = plt.pcolormesh(ds.lon, ds.lat, temp_anomaly, shading='auto', cmap='coolwarm')
plt.colorbar(im, label="Temperature Anomaly (K)")
plt.title(f"Temperature Anomaly at ~{pfull_bars[closest_idx].values:.2f} bar – Time: {last_time.values}")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```
